```python
file_in = open('pwin-Copy1.txt', 'r')
lines = file_in.readlines()
file_in.close()

lineCounter = 0
faultyPW = 0
for line in lines:
    lineCounter+=1
    info = line.split(' ')
    limits = info[0].split('-')
    letter = info[1].split(':')[0]
    password = info[2]
    print("password: " + str(lineCounter) + " " + password)

    counter = 0    
    for lim in limits:
        print("limits: " + lim, end = ' ')
    print()
    
    for c in password:
        if c == letter:
            print(c + " : " + letter)
            counter+=1
    
    if counter < int(limits[0]):
        faultyPW+=1        
    if counter > int(limits[1]):
        faultyPW+=1

print("Fucked passwords: " + str(faultyPW))
```
