```python
import os

file_in = open('pwin-Copy1.txt', 'r')
lines = file_in.readlines()

lineCounter = 0
faultyPW = 0
for line in lines:
    lineCounter+=1
    info = line.split(' ')
    limits = info[0].split('-')
    letter = info[1].split(':')[0]
    password = info[2]
    print("password: " + str(lineCounter) + " " + password)

    counter = 0
    if password[int(limits[0]) - 1] == password[int(limits[1]) - 1]:
        faultyPW+=1
        continue
    if password[int(limits[0]) - 1] != letter and password[int(limits[1]) - 1] != letter:
        faultyPW+=1
        continue

print("Fucked passwords: " + str(faultyPW))
```
