```python
import re

byr_bit=0x01
iyr_bit=0x02
eyr_bit=0x04
hgt_bit=0x08
hcl_bit=0x10
ecl_bit=0x20
pid_bit=0x40
cid_bit=0x80

file_in = open('input.txt', 'r')
file = file_in.read()
file_in.close()

# All seperate passports
passports=file.split('\n\n')
valid=0
pp_counter=0
for passport in passports:
    bitfield=0
    bitfield|=cid_bit
    data=re.split(r'\ |\n|:', passport)
    for i in range(0, len(data), 2):
        cred_id = data[i]
        try:
            if cred_id != 'pid':
                cred_data = int(data[i+1])
            else:
                cred_data = data[i+1]
        except:
            cred_data = data[i+1]

        if(cred_id == 'byr'):
            if (cred_data >= 1920 and cred_data <= 2002):
#                 print("byr:" + str(cred_data))
                bitfield|=byr_bit
        elif(cred_id == 'iyr'):
            if (cred_data >= 2010 and cred_data <= 2020):
#                 print("IYR:" + str(cred_data))
                bitfield|=iyr_bit
        elif(cred_id == 'eyr'):
            if (cred_data >= 2020 and cred_data <= 2030):
#                 print("EYR:" + str(cred_data))
                bitfield|=eyr_bit
        elif(cred_id == 'hgt'):
            if type(cred_data) is not int:
#                 print(type(cred_data))
                number = re.findall(r'\d+', cred_data)[0]
                if('in' in cred_data):
                    if (int(number) >= 59 and int(number) <= 76):
                        bitfield|=hgt_bit
                if('cm' in cred_data):
                    if (int(number) >= 150 and int(number) <= 193):
                        bitfield|=hgt_bit
        elif(cred_id == 'hcl'):
            if type(cred_data) is not int:
                if(re.fullmatch("#[0-9a-f]{6}", cred_data)):
                    bitfield|=hcl_bit
        elif(cred_id == 'ecl'):
            if("amb" in cred_data or
              "blu" in cred_data or
              "brn" in cred_data or
              "gry" in cred_data or
              "grn" in cred_data or
              "hzl" in cred_data or
              "oth" in cred_data):
                bitfield|=ecl_bit
        elif(cred_id == 'pid'):
            if(re.fullmatch("[0-9]{9}", cred_data )):
                bitfield|=pid_bit
    print("Doc#: " + str(pp_counter) + " Bitfield: " + str(hex(bitfield)))
    if bitfield is 0xFF:
        print("    VALID!")
        valid+=1
    pp_counter+=1

print("Valid: " + str(valid))  
# 114
```
